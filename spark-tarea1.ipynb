```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("Python Spark SQL Tarea 1") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()
```

```python
# conexión a Spark
sc = spark.sparkContext
```

```python
# textFile carga un archivo de texto y crea un RDD a partir del mismo
rdd_csv1 = sc.textFile("alturas_v0.csv")
rdd_csv2 = sc.textFile("alturas.csv")
```

```python
# collect recupera todos los elementos de un RDD desde los nodos del clúster de Spark
# los trae de vuelta al programa de control en una lista
rdd_altura = rdd_csv2.collect()
rdd_alturav0 = rdd_csv1.collect()
# print(rdd_altura)
# print(rdd_altura0)
```

```python
conteo1 = rdd_csv1.count()
conteo2 = rdd_csv2.count()
conteo1, conteo2
```

```python
# pregunta 1
# se aplica función lambda que realiza las siguientes operaciones:
# si i (cada elemento en el RDD) es igual a -100 se asigna None
# si i es menor que 5 se multiplica por 100, convertir metros a centímetros
# i se convierte a float
# finalmente se filtran los valores None 
rdd_corregido = rdd_csv1.map(lambda i: None if float(i) == -100 
                        else float(i) * 100 
                        if float(i) < 5 
                        else float(i)) \
                           .filter(lambda x: x is not None)
# suma de todas las alturas con sum
suma_alturas = rdd_corregido.sum()
#total de alturas con count
num_alturas = rdd_corregido.count()
#media
media = suma_alturas / num_alturas
#suma de los cuadrados de las diferencias entre cada altura y la media
suma_cuadrados = rdd_corregido.map(lambda x: (x - media) ** 2).sum()
#dividiendo la suma de cuadrados entre el número de alturas y tomando la raíz cuadrada
desviacion_tipica = (suma_cuadrados / num_alturas) ** 0.5

#media y desviación típica redondeadas a dos decimales
print("Media de alturas:", round(media,2))
print("Desviación típica de alturas:", round(desviacion_tipica,2))
```

```python
# filtrar el RDD para obtener dos RDDs separados para "H" y "M"
rdd_hombres = rdd_csv2.filter(lambda line: "H" in line.split(',')[0]).map(lambda line: line.split(',')[1])
rdd_mujeres = rdd_csv2.filter(lambda line: "M" in line.split(',')[0]).map(lambda line: line.split(',')[1])

# mostrar para verificar
print("RDD para Hombres:")
print(rdd_hombres.collect())

print("\nRDD para Mujeres:")
print(rdd_mujeres.collect())
```

```python
rdd_H = rdd_hombres.map(lambda i: None if float(i) == -100 
                        else float(i) * 100 
                        if float(i) < 5 
                        else float(i)) \
                           .filter(lambda x: x is not None)

suma_alturas1 = rdd_H.sum()
num_alturas1 = rdd_H.count()
media1 = suma_alturas1 / num_alturas1

suma_cuadrados1 = rdd_H.map(lambda x: (x - media1) ** 2).sum()
desviacion_tipica1 = (suma_cuadrados1 / num_alturas1) ** 0.5

print("Media de alturas:", round(media1,2))
print("Desviación típica de alturas:", round(desviacion_tipica1,2))
```

```python
rdd_M = rdd_mujeres.map(lambda i: None if float(i) == -100 
                        else float(i) * 100 
                        if float(i) < 5 
                        else float(i)) \
                           .filter(lambda x: x is not None)

suma_alturas2 = rdd_M.sum()
num_alturas2 = rdd_M.count()
media2 = suma_alturas2 / num_alturas2

suma_cuadrados2 = rdd_M.map(lambda x: (x - media2) ** 2).sum()
desviacion_tipica2 = (suma_cuadrados2 / num_alturas2) ** 0.5

print("Media de alturas:", round(media2,2))
print("Desviación típica de alturas:", round(desviacion_tipica2,2))
```
